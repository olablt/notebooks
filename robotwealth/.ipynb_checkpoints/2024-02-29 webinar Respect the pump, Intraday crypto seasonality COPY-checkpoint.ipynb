The big idea:

- at any point a bunch of non-random trading is happening
- some of it probably has some daily regularity ("fund starts twapping every day at a certain hour")

So I was interested in: "Do the recent returns during some time of the day, explain future returns at that time?"

i.e. if SOL performed best at 1 UTC over the last 30 days, is it more likely to keep performing best in that hour tomorrow?

```r
prices <- raw_prices %>%
  rename(
      ticker = Ticker,
      datetime = Datetime,
      open = Open,
      high = High,
      low = Low,
      close = Close,
      volume = Volume,
      quote_volume = `Quote asset volume`,
      num_trades = `Number of trades`,
      ask = `Taker buy base asset volume`,
      # bid = volume - ask,
      # delta = ask - bid
  ) %>%
  mutate (
      bid = volume - ask,
      delta = ask - bid,
  ) %>%
  select ( -`Taker buy quote asset volume`)
  
prices %>% filter(ticker == "BTC") %>% tail()
```

## Take top n by rolling 30-day dollar volume

```r
# limit universe
trading_universe_size <- 50

universe <- prices %>%
  # just get the top n by trailing 30-day volume
  group_by(ticker) %>%
  mutate(trail_volume = roll::roll_mean(quote_volume, 30*24)) %>%
  na.omit() %>%
  group_by(datetime) %>%
  mutate(
    volume_rank = row_number(-trail_volume),
    is_universe = volume_rank <= trading_universe_size
  )

head(universe)
```

## obvious general intraday seasonality

```r
# trail_volatility: This is the rolling standard deviation of the log returns,
# calculated over a window of 7*24 observations. The result is then annualized
# by multiplying by the square root of 365. This assumes that the data points are
# hourly (given the multiplication by 24), and aims to measure how much the returns
# of the asset vary over a typical week.

# vol_adjusted_returns: This represents returns adjusted for volatility, calculated
# as the returns times a factor (0.5 / lag(trail_volatility)). The adjustment uses
# the lagged (previous period's) trailing volatility, meaning it scales the returns
# based on the previously observed volatility. The constant 0.5 is used to scale
# the adjustment, and the exact choice of this constant depends on the desired
# sensitivity to volatility.

# In summary, this code transforms a financial dataset by calculating hourly log
# returns, their trailing volatility over a week, and adjusting the returns by the
# observed volatility. It also extracts the trading hour for further analysis.

returns <- universe %>%
  group_by(ticker) %>%
  arrange(datetime) %>%
  mutate(
    log_return = log(close/lag(close)),
    trail_volatility = roll::roll_sd(log_return, 7*24) * sqrt(365),
    vol_adjusted_returns = log_return * (0.5 / lag(trail_volatility)), # 7 day trailing vol to size things by
    hour = lubridate::hour(datetime)
  ) %>%
  na.omit()
```

```r
# Box plot returns by hour
returns %>%
  filter(is_universe, Datetime >= '2023-01-01') %>%
  group_by(hour) %>%
  ggplot(aes(x=hour, group=hour, y=vol_adjusted_returns)) + geom_boxplot() + ggtitle('Vol-adjusted crypto spot returns by hour of day', subtitle = '2023-2024: top 50 coins by 30 day trailing volume')
```

```r
returns %>%
  filter(is_universe, datetime >= '2023-01-01') %>%
  group_by(hour) %>%
  summarize(mean_annualized_returns_pct = mean(vol_adjusted_returns) * 365 * 12 * 100) %>%
  ggplot(aes(x=hour, y=mean_annualized_returns_pct)) + geom_point() +
    ggtitle('Mean vol-adjusted crypto spot returns by hour of day', subtitle = '2023-2024: top 50 coins by 30 day trailing volume')
```

some hint of exceess returns prior to 0 UTC and low returns immediately after - as some have anecodtally mentioned to me. Does it hold up if we look at each year?

```r
returns %>%
  mutate(year = lubridate::year(datetime)) %>%
  filter(is_universe) %>%
  group_by(hour, year) %>%
  summarize(mean_annualized_returns_pct = mean(vol_adjusted_returns) * 365 * 12 * 100) %>%
  ggplot(aes(x=hour, y=mean_annualized_returns_pct)) + geom_point() + facet_wrap(~year, scales='free') +
    ggtitle('Mean vol-adjusted crypto spot returns by hour of day', subtitle = '2023-2024: top 50 coins by 30 day trailing volume')
```

## Looking for Persistence of Seasonality Patterns

Is there any persistence to what tickers tend to do better at certain times of the day?

For each ticker, rank its risk-adjsuted returns for each hour over some lookback. Is that rank correlated to future returns over that hour?

Here we

-    calculate mean-vol adjusted returns for each ticker each hour
-    calculate the rolling 30 day returns for each ticker each hour
-    for each ticker, for each day, rank each of the hourse ** 1 = worst performing hour over the last 30 days ** 2 = best performing hour over the last 30 hours

Then look to see if that (lagged) rank predicts next days vol-adjusted returns for that ticker on that hour.


```r
lookback <- 30 # in days

hour_persistence <- returns %>%
  mutate(date = as.Date(datetime)) %>%
  group_by(ticker, hour) %>%
  arrange(datetime) %>%
  mutate(trail_perf = roll::roll_mean(vol_adjusted_returns, lookback)) %>%
  na.omit() %>%
  # now calculate the rank for each ticker
  group_by(ticker, date) %>%
  mutate(rank = row_number(trail_perf)) %>%
  # Now does that lagged rank predict returns
  group_by(ticker, hour) %>%
  arrange(datetime) %>%
  mutate(lagged_rank = lag(rank)) %>%
  na.omit()
```

```r
hour_persistence %>%
  filter(is_universe) %>%
  group_by(lagged_rank) %>%
  summarize(meanreturns = mean(vol_adjusted_returns) * 365 * 12 * 100) %>%
  ggplot(aes(x=lagged_rank, y=meanreturns)) + geom_point() + ggtitle('Mean vol-adjusted annualized returns by ranked hour of the day over past 30 days')
```



strong effect. let's see if it exists so strongly in the most obvious ticker, bitcoin


```r
hour_persistence %>%
  filter(Ticker == 'BTC') %>%
  group_by(lagged_rank) %>%
  summarize(meanreturns = mean(vol_adjusted_returns) * 365 * 12 * 100) %>%
  ggplot(aes(x=lagged_rank, y=meanreturns)) + geom_point() + ggtitle('Mean vol-adjusted annualized returns by ranked hour of the day over past 30 days - BTC')
```



not so much.

we'll do a factor plot... and rank each ticker/hour long/short in proportion to how well it did in the last 30 days.


```r
hour_persistence %>%
  filter(date >= '2020-01-01') %>%
  filter(is_universe) %>%
  mutate(date = as.Date(Datetime)) %>%
  mutate(weight = lagged_rank - 12.5) %>%
  group_by(date) %>%
  summarize(mean_returns = mean(vol_adjusted_returns * weight)) %>%
  arrange(date) %>%
  mutate(cumreturns = cumsum(mean_returns)) %>%
  ggplot(aes(x=date, y=cumreturns)) + geom_line() +
  ggtitle('Hold each coin in proportion to how well it did that hour over the last 30 days', subtitle = 'Top 50 coins by 30 day trailing volume')
```



looks pretty good.

here's something that looks more like a trading strategy. simply long the each ticker in its best hour and short it in its worst hour.


```r
hour_persistence %>%
  filter(date >= '2020-01-01') %>%
  filter(is_universe) %>%
  mutate(date = as.Date(Datetime)) %>%
  filter(lagged_rank %in% c(1,24)) %>%
  mutate(weight = case_when(lagged_rank < 12 ~ -1, lagged_rank >= 12 ~ 1)) %>%
  group_by(date) %>%
  summarize(mean_returns = mean(vol_adjusted_returns * weight)) %>%
  arrange(date) %>%
  mutate(cumreturns = cumsum(mean_returns)) %>%
  ggplot(aes(x=date, y=cumreturns)) + geom_line() +
  ggtitle('Long(short) each coin in its best(worst) hour over the last 30 days', subtitle = 'Top 50 spot coins by 30 day trailing volume')
```



as a sense check, what if we approached this from the other direction?

For each hour, rank the assets. do the assets that did best in a given hour tend to outperform.


```r
lookback <- 30 # in days

hour_asset_persistence <- returns %>%
  mutate(date = as.Date(Datetime)) %>%
  group_by(Ticker, hour) %>%
  arrange(Datetime) %>%
  mutate(trail_perf = roll::roll_mean(vol_adjusted_returns, lookback)) %>%
  na.omit() %>%
  filter(is_universe) %>%
  # now calculate the rank of each asset for each ticker
  group_by(hour, date) %>%
  mutate(rank = row_number(trail_perf)) %>%
  # Now does that lagged rank predict returns
  group_by(Ticker, hour) %>%
  arrange(Datetime) %>%
  mutate(lagged_rank = lag(rank)) %>%
  na.omit()
```

```r
hour_asset_persistence %>%
  filter(is_universe) %>%
  group_by(lagged_rank) %>%
  summarize(meanreturns = mean(vol_adjusted_returns) * 365 * 12 * 100) %>%
  ggplot(aes(x=lagged_rank, y=meanreturns)) + geom_point() + ggtitle('Mean vol-adjusted annualized returns by ranked asset for each hour of the day over past 30 days')
```



we also see a clear effect there, which is comforting.

this is probably untradeable like this. let's break the day down into bigger chunks. say 6 hour chunks.


```r
lookback <- 30 # in days

fourhour_persistence <- returns %>%
  mutate(date = as.Date(Datetime)) %>%
  mutate(hour_chunk = ceiling((hour+1) * (6/24))) %>% # four six-hour chunks
  group_by(date, Ticker, hour_chunk) %>%
  summarize(vol_adjusted_returns = sum(vol_adjusted_returns), is_universe = last(is_universe)) %>%
  group_by(Ticker, hour_chunk) %>%
  arrange(date) %>%
  mutate(trail_perf = roll::roll_mean(vol_adjusted_returns, lookback)) %>%
  na.omit() %>%
  # now calculate the rank for each ticker
  group_by(Ticker, date) %>%
  mutate(rank = row_number(trail_perf)) %>%
  # Now does that lagged rank predict returns
  group_by(Ticker, hour_chunk) %>%
  arrange(date) %>%
  mutate(lagged_rank = lag(rank)) %>%
  na.omit()

fourhour_persistence %>%
  filter(is_universe) %>%
  group_by(lagged_rank) %>%
  summarize(meanreturns = mean(vol_adjusted_returns) * 365 * 12 * 100) %>%
  ggplot(aes(x=lagged_rank, y=meanreturns)) + geom_point() + ggtitle('Mean vol-adjusted annualized returns by ranked four-hour period over past 30 days')
```



still see an effect


```r
hour_persistence %>%
  filter(date >= '2020-01-01') %>%
  filter(is_universe) %>%
  filter(lagged_rank %in% c(1,6)) %>%
  mutate(weight = case_when(lagged_rank < 3.5 ~ -1, lagged_rank >= 3.5 ~ 1)) %>%
  group_by(date) %>%
  summarize(mean_returns = mean(vol_adjusted_returns * weight)) %>%
  arrange(date) %>%
  mutate(cumreturns = cumsum(mean_returns)) %>%
  ggplot(aes(x=date, y=cumreturns)) + geom_line() +
  ggtitle('Long(short) each coin in its best(worst) four hours over the last 30 days', subtitle = 'Top 50 spot coins by 30 day trailing volume')
```

still looks pretty good huh?

## Do we see day of week seasonal effects like this too?



Let's look at day of week seasonal effects in the same way.


```r
daily <- returns %>%
  mutate(date = as.Date(Datetime)) %>%
  group_by(Ticker, date) %>%
  arrange(Datetime) %>%
  summarize(
    Close = last(Close),
    volume = sum(`quote_volume`),
    volume_rank = last(volume_rank),
    is_universe = last(is_universe),
    vol_adjusted_returns = sum(vol_adjusted_returns)
  )
```

```r
lookback <- 50 # in weeks

dow_persistence <- daily %>%
  mutate(
    dow = lubridate::wday(date),
    sow = lubridate::floor_date(date, 'week')
  ) %>%
  group_by(Ticker, dow) %>%
  arrange(date) %>%
  mutate(trail_perf = roll::roll_mean(vol_adjusted_returns, lookback)) %>%
  na.omit() %>%
  # now calculate the rank for each ticker
  group_by(Ticker, sow) %>%
  mutate(rank = row_number(trail_perf)) %>%
  # Now does that lagged rank predict returns
  group_by(Ticker, dow) %>%
  arrange(date) %>%
  mutate(lagged_rank = lag(rank)) %>%
  na.omit()

dow_persistence %>%
  filter(date >= '2020-01-01') %>%
  filter(is_universe) %>%
  group_by(lagged_rank) %>%
  summarize(meanreturns = mean(vol_adjusted_returns) * 365 * 100) %>%
  ggplot(aes(x=lagged_rank, y=meanreturns)) + geom_point() + ggtitle('Mean vol-adjusted annualized returns by ranked day of week over past 50 weeks')
```



Not much evidence of a persistence effect if we look back 50 weeks.

### Set Returns

```r
returns <- universe %>%
  group_by(ticker) %>%
  filter(is_universe) %>%
  arrange(datetime) %>%
  mutate(
    hour = lubridate::hour(datetime),
    # trail_volatility = roll::roll_sd(returns, 7*24) * sqrt(365),
    # vol_adjusted_returns = returns * (0.5 / lag(trail_volatility)), # 7 day trailing vol to size things by
    # returns
    simple_return = (close / lag(close) - 1),
    log_return = log(close / lag(close)),
    future_simple_return = (lead(close) / close - 1),
    future_log_return = log(lead(close) / close),
  ) %>%
  na.omit()
# returns %>% head()
```

### Define Breakout, Momo, Rate Features

```r

rolling_periods_since_high_72 <- purrr::possibly(
  tibbletime::rollify(
    function(x) {
      idx_of_high <- which.max(x)
      periods_since_high <- length(x) - idx_of_high
      periods_since_high
    },
    window = 72, na_value = NA),
  otherwise = NA
)

# Function to scale the periods since high into a range from -9.5 to 9.5
scale_periods_since_high <- function(periods_since_high, total_periods) {
  # Calculate which decile the period falls into (1 to 10)
  decile <- findInterval(periods_since_high, seq(0, total_periods, length.out = 11), rightmost.closed = TRUE)
  # Scale from 1-10 to -9.5 to 9.5
  scaled_value <- decile - 5.5
  return(scaled_value)
}
rolling_periods_since_high <- function(close_prices, window_size) {
  purrr::possibly(
    tibbletime::rollify(
      function(x) {
        idx_of_high <- which.max(x)
        periods_since_high <- length(x) - idx_of_high
        # Call the scaling function to transform periods_since_high to a range from -9.5 to 9.5
        scale_periods_since_high(periods_since_high, length(x))
      },
      window = window_size, 
      na_value = NA
    ),
    otherwise = NA
  )(close_prices)
}

features <- returns %>%
  group_by(ticker) %>%
  # group_by(datetime) %>%
  arrange(datetime) %>%
  mutate(
    # we won't lag features here because we're using forward returns
    # breakout = 71.5 - rolling_hours_since_high_72(close),  # puts this feature on a scale -9.5 to +9.5 (9.5 = high was today)
    breakout = rolling_hours_since_high_72(close),  # puts this feature on a scale -9.5 to +9.5 (9.5 = high was today)
    # breakout = rolling_periods_since_high(close, 92),
    momo = (close - lag(close, 10)) / lag(close, 10), # standard momentum calculation is usually the percentage change over a period
    rate = close - lag(close, 10)/close
  ) %>%
  ungroup() %>%
  na.omit()

head(features)
```

```r
features %>%
  filter(is_universe) %>%
  pivot_longer(c(breakout, momo, rate, delta), names_to = "feature") %>%
  ggplot(aes(x = value, colour = feature)) +
  geom_density() +
  facet_wrap(~feature, scales = "free")
```

### Scale Features to Zscores and Deciles

```r
features_scaled <- features %>%
  filter(is_universe) %>%
  group_by(datetime) %>%
  # group_by(ticker) %>%
  mutate(
      demeaned_future_simple_returns = future_simple_return - mean(future_simple_return),
      
    # # Calculating the Median and IQR for robust scaling
    # delta_median = median(delta, na.rm = TRUE),
    # delta_iqr = IQR(delta, na.rm = TRUE),
    # momo_median = median(momo, na.rm = TRUE),
    # momo_iqr = IQR(momo, na.rm = TRUE),
    # rate_median = median(rate, na.rm = TRUE),
    # rate_iqr = IQR(rate, na.rm = TRUE),
    # # Robust Z-scores
    # delta_zscore = (delta - delta_median) / delta_iqr,
    # momo_zscore = (momo - momo_median) / momo_iqr,
    # rate_zscore = (rate - rate_median) / rate_iqr,
      
    delta_zscore = -(delta - mean(delta, na.rm = TRUE)) / sd(delta, na.rm = TRUE),
    momo_zscore = -(momo - mean(momo, na.rm = TRUE)) / sd(momo, na.rm = TRUE),
    rate_zscore = -(rate - mean(rate, na.rm = TRUE)) / sd(rate, na.rm = TRUE),
    # Decile scaling
    delta_decile = 5.5 - ntile(delta, 10), # scaling to values from -4.5 to 4.5
    momo_decile = 5.5 - ntile(momo, 10), # scaling to values from -4.5 to 4.5
    rate_decile = 5.5 - ntile(rate, 10), # scaling to values from -4.5 to 4.5
  ) %>%
  na.omit() %>%
  ungroup()

# features_scaled %>%
#   pivot_longer(c("delta_zscore", "momo_zscore", "rate_zscore", "delta_decile", "momo_decile", "rate_decile"), names_to = "feature") %>%
#   ggplot(aes(x = value, colour = feature)) +
#   geom_density() +
#   facet_wrap(~feature, scales = "free") 
features_scaled %>%
  pivot_longer(c("delta_zscore", "momo_zscore", "rate_zscore", "delta_decile", "momo_decile", "rate_decile"), names_to = "feature") %>%
  ggplot(aes(x = value, colour = feature)) +
  geom_density() +
  facet_wrap(~feature, scales = "free") +
  labs(title = "Density of Scaled Features", x = "Scaled Value", y = "Density")
```

### Information coefficient
The information coefficient is simply the correlation of the feature with next day returns. It provides a single metric for quantifying the strength of the signal. We can calculate it for time-series returns and cross-sectional returns:

```r
features_scaled %>%
  # pivot_longer(c(momo_decile), names_to = "feature") %>%
  pivot_longer(c(breakout, 
                 momo_zscore, momo_decile, 
                 delta_zscore, delta_decile, 
                 rate_zscore, rate_decile), names_to = "feature") %>%
  group_by(feature) %>%
  summarize(IC = cor(value, future_log_return))  %>%
  ggplot(aes(x = factor(feature, levels = 
    c('breakout', 
      'momo_zscore', 'momo_decile', 
      'delta_zscore', 'delta_decile', 
      'rate_zscore', 'rate_decile')), y = IC)) +
  geom_bar(stat = "identity") +
  labs(
    title = "Correlation of Feature Z-Scores and Deciles with Future Returns",
    y = "Information Coefficient (IC)",
    x = "Feature"
  )
```

### Decay characteristics

Next, it’s useful to understand how quickly your signals decay. Signals that decay quickly need to be acted upon without delay, while those that decay slower are more forgiving to execute.

We’ll look at relative return IC for our cross-sectional features and the time-series return IC for our breakout feature.

## What Is LogReturn

### Absolute change in price

```r
prices %>%
  filter(ticker == "BTC") %>%
  mutate (
      change = close - lag(close, 1),
  ) %>%
  na.omit() %>%
  ggplot(aes(x=datetime, y=cumsum(change))) + geom_line()

# The expression returns = close - lag(close, 1) calculates the absolute change
# in price from one period to the next, not the return as a percentage. While it
# does give you the difference in price, it doesn't take into account the scale
# of the investment.

# In financial analysis, returns are typically expressed as a percentage
# because they provide a scale-independent measure of investment performance.

# This is important because:
# - Comparability: Percent returns allow for the comparison of investments
# of different sizes. An absolute return of $10 on a $100 investment is a
# 10% return, which is quite good. However, a $10 return on a $10,000 investment
# is only a 0.1% return, which is much less impressive.
# - Normalization: Percentage returns normalize the gains or losses relative to
# the size of the investment. This normalization allows investors to make more
# informed decisions based on the relative profitability of investments, rather
# than the absolute dollar amounts, which can be misleading.
# - Compound Interest: When considering compound interest, which is the interest
# on the initial principal and also on the accumulated interest of previous
# periods, percentage returns provide a more accurate and meaningful figure for
# growth over time.

# For these reasons, when evaluating investment performance, financial analysts
# usually prefer using percentage returns rather than absolute price changes.
# This approach is standard across the industry, as it provides a clear picture
# of investment efficiency.
```

### Simple Returns

```r
# claculate simple returns and draw  simple cumsum
prices %>%
  filter(ticker == "BTC") %>%
  mutate (
      # returns = close - lag(close, 1),
      returns = close / lag(close) - 1,
  ) %>%
  na.omit() %>%
  ggplot(aes(x=datetime, y=cumsum(returns))) +
  geom_line()

# - When you divide the current price by the previous price, you are 
# essentially calculating how much the current price is as a multiple 
# of the previous price. Subtracting 1 from this ratio gives you the 
# percentage change from the previous price to the current price.
# - Not subtracting 1 would give you the total value now for every 
# dollar invested before, including the original investment, rather 
# than the net gain or loss as a percentage. Subtracting 1 gives you 
# just the gain or loss, which is generally what's most relevant for 
# analyzing returns over time.
```

### Log Returns

```r
# calculate log returns and draw a cumsum
prices %>%
  filter(ticker == "BTC") %>%
   mutate (
       log_returns = log(close / lag(close, 1)),
       future_log_returns = lead(log_returns, 1),
   ) %>%
   na.omit() %>%
   # ggplot(aes(x=datetime, y=cumsum(future_log_returns))) +
   ggplot(aes(x=datetime, y=cumsum(log_returns))) +
   geom_line()

# - Simple returns are not time additive: The sum of simple returns over time does not give you the total return.
# - Compounding: Simple returns compound by multiplication, while log returns compound by addition.
```

### Total Percentage Change

```r
# calculate log returns and draw Total Percentage Change a cumsum
prices %>%
  filter(ticker == "BTC") %>%
   mutate (
       log_returns = log(close / lag(close, 1)),
       future_log_returns = lead(log_returns, 1),
   ) %>%
   na.omit() %>%
   # get the total compounded return as a percentage
   ggplot(aes(x=datetime, y=exp(cumsum(log_returns)-1))) +
   # ggplot(aes(x=datetime, y=exp(cumsum(future_log_returns)-1))) +
   geom_line()

# - To get the Total Percentage Change from cumulative log returns, you need to exponentiate the final value
# - Simple returns are not time additive: The sum of simple returns over time does not give you the total return.
# - Compounding: Simple returns compound by multiplication, while log returns compound by addition.
```
