```r
# Read CSV file
df <- read_csv("neurotrader888/IntramarketDifference/bybit_prices_2017-2024.csv")
```

# Correlation

```r
# Assuming df is already loaded and manipulated as per your code
data <- df %>%
  filter(symbol %in% c("BTCUSDT", "ETHUSDT"),
         date >= as.POSIXct("2021-03-15 00:00:00", tz = "UTC")) %>%
  group_by(symbol) %>%
  mutate(diff = log(close) - lag(log(close))) %>%
  ungroup() # Ensure data is not grouped for further operations

# Separate data for BTC and ETH
btc_diff <- data %>% filter(symbol == "BTCUSDT") %>% select(date, diff) %>% na.omit()
eth_diff <- data %>% filter(symbol == "ETHUSDT") %>% select(date, diff) %>% na.omit()

# Aligning the dates might be necessary if the trading dates do not completely overlap
# This code assumes dates are aligned. If they are not, you'd need to join by date before computing correlation
correlation <- cor(btc_diff$diff, eth_diff$diff)

print(paste("Correlation:", correlation))
```

# Chart Two Symbols

```r
# Compute the scaling factor outside of the chain
btc_range <- data %>%
  filter(symbol == "BTCUSDT") %>%
  summarize(range = max(close) - min(close)) %>%
  pull(range)

eth_range <- data %>%
  filter(symbol == "ETHUSDT") %>%
  summarize(range = max(close) - min(close)) %>%
  pull(range)

print(btc_range)
print(eth_range)

scaling_factor <- btc_range / eth_range

# Now we pipe the data frame into ggplot in a single chain
df %>%
  # Filter out data to only include BTC and ETH from a certain date
  filter(symbol %in% c("BTCUSDT", "ETHUSDT"), date >= as.Date("2021-03-15")) %>%
  # Plot
  ggplot(aes(x = date)) +
  geom_line(data = . %>% filter(symbol == "BTCUSDT"), aes(y = close), color = "blue") +
  geom_line(data = . %>% filter(symbol == "ETHUSDT"), aes(y = close * scaling_factor), color = "red") +
  scale_y_continuous(
    name = "BTC Close",
    sec.axis = sec_axis(~ . / scaling_factor, name = "ETH Close")
  ) +
  theme_minimal() +
  ggtitle('Comparison of BTC and ETH Close Prices', subtitle = 'Independently Scaled Y-Axes for BTC and ETH')
```

```r
lookback <- 24  # Period for moving average
atr_lookback <- 168  # Period for ATR
threshold <- 0.25

library(roll)

# cmma values
cmma <- df %>%
  filter(symbol %in% c("BTCUSDT", "ETHUSDT"), date >= as.Date("2021-03-15")) %>%
  arrange(symbol, date) %>%  # Ensure data is sorted
  group_by(symbol) %>%
#   arrange(date) %>%
  mutate(
    # Calculate the moving average using rollmean from the zoo package
    moving_avg = rollmean(close, lookback, fill = NA, align = 'right'),
    # Calculate True Range (TR)
    tr = pmax(high - low, abs(high - lag(close)), abs(low - lag(close)), na.rm = TRUE),
    # Calculate the ATR using rollmean on the TR
    atr = rollmean(tr, atr_lookback, fill = NA, align = 'right'),
    # Calculate the new indicator (ind)
    ind = (close - moving_avg) / (atr * sqrt(lookback)),
    diff = log(close) - lag(log(close)),
    next_return = lag(diff)
  ) %>%
  na.omit() %>%
  ungroup()  # Ungroup for further operations

# TODO: return, log_return and trading_signals, cumsum
cmma
```

```r
generate_signals <- function(ind, threshold) {
  signal <- numeric(length(ind))
  position <- 0

  for (i in seq_along(ind)) {
    if (ind[i] > threshold) {
      position <- 1
    } else if (ind[i] < -threshold) {
      position <- -1
    }

    if (position == 1 && ind[i] <= 0) {
      position <- 0
    } else if (position == -1 && ind[i] >= 0) {
      position <- 0
    }

    signal[i] <- position
  }

  return(signal)
}


# Assuming `cmma` has an `ind` column for each symbol
cmma <- cmma %>%
  group_by(symbol) %>%
  mutate(sig = generate_signals(ind, threshold)) %>%
  ungroup()
# cmma

# Assuming `df` is your data frame with the `sig` and `next_return` columns calculated
entries <- cmma %>%
  mutate(
    # Mark trade entries and exits for long positions
    long_entry = lag(sig) == 0 & sig == 1,
    long_exit = lag(sig) == 1 & sig == 0,
    # Mark trade entries and exits for short positions
    short_entry = lag(sig) == 0 & sig == -1,
    short_exit = lag(sig) == -1 & sig == 0
  )

# Initialize columns for trade returns
returns <- entries %>%
  mutate(
    long_return = ifelse(long_exit, next_return, NA),
    short_return = ifelse(short_exit, -next_return, NA) # Note the negation for short trades
  )

# Filter out the rows where trades are closed (exits) to calculate cumulative returns
trade_returns <- returns %>%
  filter(!is.na(long_return) | !is.na(short_return)) %>%
  mutate(return = coalesce(long_return, short_return))

# Calculate cumulative returns
trade_returns <- trade_returns %>%
  mutate(cumulative_return = cumsum(return))


ggplot(trade_returns, aes(x = date, y = cumulative_return)) +
  geom_line() +
  theme_minimal() +
  labs(title = "Cumulative Returns from Trading Strategy",
       x = "Date",
       y = "Cumulative Return")


# get_trades_from_signal <- function(data) {
#   trades <- data.frame(entry_time = numeric(), entry_price = numeric(),
#                        exit_time = numeric(), exit_price = numeric(),
#                        return = numeric(), type = numeric())

#   current_trade <- NULL
#   for (i in 2:nrow(data)) {
#     if (data$sig[i-1] == 0 && data$sig[i] != 0) {
#       # Trade entry
#       current_trade <- list(entry_time = data$date[i], entry_price = data$close[i], type = data$sig[i])
#     } else if (!is.null(current_trade) && data$sig[i] == 0) {
#       # Trade exit
#       current_trade$exit_time <- data$date[i]
#       current_trade$exit_price <- data$close[i]
#       current_trade$return <- ifelse(current_trade$type == 1,
#                                      (current_trade$exit_price - current_trade$entry_price) / current_trade$entry_price,
#                                      (current_trade$entry_price - current_trade$exit_price) / current_trade$entry_price)
#       trades <- rbind(trades, unlist(current_trade))
#       current_trade <- NULL
#     }
#   }

#   return(trades)
# }

# get_trades_from_signal(cmma)
```

```r
library(xts)
cmma <- function(ohlc, lookback, atr_lookback = 168) {
      # Ensure ohlc is an xts object for TTR compatibility
  if (!is.xts(ohlc)) {
    ohlc_xts <- xts(ohlc[, c("High", "Low", "Close")], order.by=ohlc$OpenTime)
  } else {
    ohlc_xts <- ohlc
  }

  atr <- ATR(HLC = ohlc_xts, n = atr_lookback)[, "atr"]
  ma <- runMean(ohlc_xts[, "Close"], n = lookback)
  ind <- (ohlc_xts[, "Close"] - ma) / (atr * sqrt(lookback))

  # Convert back to a regular vector if needed
  return(coredata(ind))

#   atr <- ATR(HLC = ohlc[, c("High", "Low", "Close")], n = atr_lookback)$atr
#   ma <- rollapply(ohlc$Close, width = lookback, FUN = mean, align = "right", fill = NA)
#   ind <- (ohlc$Close - ma) / (atr * sqrt(lookback))
#   return(ind)
}

# Prepare BTC and ETH data frames
btc_data <- filtered %>%
  filter(Symbol == 'BTCUSDT')

eth_data <- filtered %>%
  filter(Symbol == 'ETHUSDT')

# Assuming the presence of high, low, close columns in btc_data and eth_data
# Example usage of the cmma function
lookback <- 24
atr_lookback <- 168
threshold <- 0.25

btc_data$cmma <- cmma(btc_data, lookback, atr_lookback)
eth_data$cmma <- cmma(eth_data, lookback, atr_lookback)

# Calculate the signal for ETH based on the difference between ETH's and BTC's CMMA
eth_data$sig <- ifelse(eth_data$cmma - btc_data$cmma > threshold, 1, ifelse(eth_data$cmma - btc_data$cmma < -threshold, -1, 0))

# Calculate next returns for ETH
eth_data$next_return <- c(diff(log(eth_data$Close)), NA)

# Calculate cumulative returns based on signals
eth_data$signal_return <- lag(eth_data$sig) * eth_data$next_return
# eth_data$cum_return <- cumsum(eth_data$signal_return, na.rm = TRUE)
# Replace NA values in 'signal_return' with 0 before cumulative sum calculation
eth_data$signal_return[is.na(eth_data$signal_return)] <- 0

# Now, calculate the cumulative returns without the need for na.rm = TRUE
eth_data$cum_return <- cumsum(eth_data$signal_return)

# Removing NA values for plotting
eth_data <- na.omit(eth_data)
```
